```python
from scipy import signal as sg
import matplotlib.pyplot as plt
```

```python
freq = 2
amp = 2
time = np.linspace(0, 2, 1000)
```

```python
signal1 = amp*np.sin(2*np.pi*freq*time)
```

```python
print(signal1)
```

```python
print(signal1[1:0])
```

```python
print(signal1[1:])
```

```python
np.savetxt('sinewave.txt',signal1, fmt='%.5f')
```

```python
plt.plot(time, signal1)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.show()
```
